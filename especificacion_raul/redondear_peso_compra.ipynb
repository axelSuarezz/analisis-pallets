```python

df['peso_de_compra_%'] = np.ceil(df['peso_de_compra'] * 100).astype(int)
```

```python
# Paso 1: Crear columna 'espacio', por ejemplo con valor fijo 1 por ahora
df['espacio'] = 2

# Paso 2: Crear 'espacio_variable_asignado'
df['espacio_variable_asignado'] = df['peso_de_compra_%'] * df['espacio']
```

```python
espacio_fijo = 10

df['espacio_total_asignado'] = espacio_fijo + df['espacio_variable_asignado']
# 
```
