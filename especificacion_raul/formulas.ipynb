```python
# PESO COMPRA
# Asegurar columna de mes tipo '2025-04'
df['mes'] = pd.to_datetime(df['fecha']).dt.strftime('%Y-%m')

# Total mensual por socio
pallets_por_socio_mes = df.groupby(['socio', 'mes'])['pallets_iniciales'].transform('sum')

# Total mensual general (de todos los socios)
pallets_totales_mes = df.groupby('mes')['pallets_iniciales'].transform('sum')

# Peso de compra
df['peso_de_compra'] = pallets_por_socio_mes / pallets_totales_mes
```

```python
# Espacio fijo
espacio_fijo = 10

# Espacio variable mensual (suponiendo 90 para todos los meses)
espacio_variable_mensual = 90

# Espacio total asignado
df['espacio_total_asignado'] = espacio_fijo + (df['peso_de_compra'] * espacio_variable_mensual)
```

```python
# ESPACIO TOTAL ASIGNADO POR MES Y SOCIO
# Pallets ingresados por socio y mes
pallets_por_socio_mes = df.groupby(['socio', 'mes'])['pallets_iniciales'].sum().reset_index(name='pallets_socio_mes')

# Total mensual
total_mes = df.groupby('mes')['pallets_iniciales'].sum().reset_index(name='total_mes')

# Merge
df_pesos = pallets_por_socio_mes.merge(total_mes, on='mes')

# Peso de compra mensual por socio
df_pesos['peso_de_compra'] = df_pesos['pallets_socio_mes'] / df_pesos['total_mes']

# Ahora sí, aplicar espacio total asignado (mensual)
espacio_fijo = 10
espacio_variable = 90  # por mes, constante por ahora

df_pesos['espacio_total_asignado'] = espacio_fijo + (df_pesos['peso_de_compra'] * espacio_variable)
```

```python
df = df.merge(df_pesos[['socio', 'mes', 'espacio_total_asignado']], on=['socio', 'mes'], how='left')
```

```python
# INDICE DE ROTACIÓN MENSUAL
# 1. Total de pallets despachados por socio y mes
despachos_mensuales = df.groupby(['socio', 'mes'])['cantidad_pallets_sacados'].sum().reset_index(name='total_despachado')

# 2. Promedio de stock diario por socio y mes
promedio_stock_mensual = df.groupby(['socio', 'mes'])['stock_ayer'].mean().reset_index(name='promedio_stock')

# 3. Unir ambos
df_rotacion = despachos_mensuales.merge(promedio_stock_mensual, on=['socio', 'mes'])

# 4. Calcular índice de rotación mensual
df_rotacion['indice_rotacion_mensual'] = df_rotacion['total_despachado'] / df_rotacion['promedio_stock']

def clasificar_rotacion(x):
    if x >= 1.0:
        return 'alta'
    elif x >= 0.5:
        return 'media'
    elif x > 0:
        return 'baja'
    else:
        return 'sin rotación'

df_rotacion['nivel_rotacion'] = df_rotacion['indice_rotacion_mensual'].apply(clasificar_rotacion)


```

```python
df = df.merge(df_rotacion[['socio', 'mes', 'indice_rotacion_mensual']], on=['socio', 'mes'], how='left')
```

```python
df.to_excel ('prueba.xlsx', index=False)
```
