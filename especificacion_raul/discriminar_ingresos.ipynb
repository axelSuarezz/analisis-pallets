```python
import pandas as pd

ruta = 'C:/Users/Sistemas/Desktop/pallets_cruzados_nuevo/cruce_2025/copia_de df_completo.xlsx'

df = pd.read_excel (ruta)

df.head()
```

```python
# Crear columna 'mes' si no estÃ¡
df['mes'] = pd.to_datetime(df['fecha']).dt.strftime('%Y-%m')

# Filtrar solo ingresos positivos
df_pos = df[df['ingresos_estimados'] > 0].copy()

# Agrupar: suma de ingresos positivos por socio y mes
ingresos_socio_mes = (
    df_pos.groupby(['socio', 'mes'])['ingresos_estimados']
    .sum()
    .rename('ingresos_por_socio_mes')
)

# Agrupar: suma total mensual de ingresos positivos
ingresos_total_mes = (
    df_pos.groupby('mes')['ingresos_estimados']
    .sum()
    .rename('ingresos_totales_mes')
)

# Unir los valores al dataframe original
df = df.merge(ingresos_socio_mes, on=['socio', 'mes'], how='left')
df = df.merge(ingresos_total_mes, on='mes', how='left')

# Calcular peso de compra
df['peso_de_compra'] = df['ingresos_por_socio_mes'] / df['ingresos_totales_mes']
```

```python
df.to_excel ('prueba.xlsx', index=False)
```
