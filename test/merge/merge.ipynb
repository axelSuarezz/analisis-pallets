```python
import pandas as pd

# Carga de archivos
archivo_base = pd.read_excel("espacio_total_asignado_final.xlsx")
archivo_mayo = pd.read_excel("mayo_cotron.xlsx")

# Mostrar info básica de ambos
print("Archivo BASE")
display(archivo_base.head())
print("\nColumnas:")
print(archivo_base.columns)
print("\nTipos de datos:")
print(archivo_base.dtypes)

print("\n" + "="*80 + "\n")

print("Archivo MAYO")
display(archivo_mayo.head())
print("\nColumnas:")
print(archivo_mayo.columns)
print("\nTipos de datos:")
print(archivo_mayo.dtypes)
```

```python

import shutil

# Crear backup del archivo original
shutil.copyfile("espacio_total_asignado_final.xlsx", "espacio_total_asignado_backup.xlsx")

# Leer los datos nuevos de mayo
df_mayo = pd.read_excel("mayo_cotron.xlsx", sheet_name="Hoja1")

# Leer el archivo total actual
df_total = pd.read_excel("espacio_total_asignado_final.xlsx", sheet_name="Sheet1")

# Asegurar que la columna 'fecha' sea datetime
df_total['fecha'] = pd.to_datetime(df_total['fecha'])
df_mayo['fecha'] = pd.to_datetime(df_mayo['fecha'])

# Filtrar datos que NO son de mayo
df_sin_mayo = df_total[df_total['fecha'].dt.month != 5]

# Unir los datos sin mayo con los nuevos de mayo
df_actualizado = pd.concat([df_sin_mayo, df_mayo], ignore_index=True)

# Ordenar por fecha y socio
df_actualizado = df_actualizado.sort_values(by=['fecha', 'socio'])

# Sobrescribir el archivo original
with pd.ExcelWriter("espacio_total_asignado_final.xlsx", engine='openpyxl', mode='w') as writer:
    df_actualizado.to_excel(writer, index=False, sheet_name="Sheet1")

```

```python
import pandas as pd
import shutil

# Paso 0: Restaurar archivo original desde backup (opcional)
shutil.copyfile("espacio_total_asignado_backup.xlsx", "espacio_total_asignado_final.xlsx")

# Paso 1: Leer archivos Excel
df_total = pd.read_excel("espacio_total_asignado_final.xlsx", sheet_name="Sheet1")
df_mayo = pd.read_excel("mayo_cotron.xlsx", sheet_name="Hoja1")

# Paso 2: Asegurar que 'fecha' sea datetime
df_total['fecha'] = pd.to_datetime(df_total['fecha'])
df_mayo['fecha'] = pd.to_datetime(df_mayo['fecha'])

# Paso 3: Eliminar duplicados en df_mayo (por si acaso)
df_mayo = df_mayo.drop_duplicates()

# Paso 4: Filtrar datos que NO sean de mayo en el archivo total
df_sin_mayo = df_total[df_total['fecha'].dt.month != 5]

# Paso 5: Concatenar datos sin mayo + datos nuevos de mayo
df_actualizado = pd.concat([df_sin_mayo, df_mayo], ignore_index=True)

# Paso 6: Ordenar por 'fecha' y 'socio'
df_actualizado = df_actualizado.sort_values(by=['fecha', 'socio'])

# Paso 7: Guardar archivo actualizado sobrescribiendo
with pd.ExcelWriter("espacio_total_asignado_final.xlsx", engine='openpyxl', mode='w') as writer:
    df_actualizado.to_excel(writer, index=False, sheet_name="Sheet1")

# Paso 8: Validación de socio "grazosi mirtha" en mayo
df_mayo_total = df_actualizado[df_actualizado['fecha'].dt.month == 5]

mirtha_filas = df_mayo_total[df_mayo_total['socio'].str.lower().str.strip() == "grazosi mirtha"]

print("Registros para 'GRAZIOSI MIRTHA' en mayo:")
print(mirtha_filas[['socio', 'fecha']])
```
